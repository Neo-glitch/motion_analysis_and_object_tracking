# Mini Project # 10 - Object Tracking

OpenCV 2.7.X only

```python
import cv2
import numpy as np

# code that initializes the webcam
cap = cv2.VideoCapture(0)

# defines the color purple range in HSV
lower_purple = np.array([130, 50, 90])
upper_purple = np.array([170, 255, 255])

# code that creates an empty array that stores our points
points = []

# get default cam window size
ret, frame = cap.read()
Height, Width = frame.shape[:2]
frame_count = 0

while 1:
    
    # code that captures the webcam frame
    ret, frame = cap.read()
    hsv_img = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    # code that threshold the hsv images to track only object in  purle colors
    mask = cv2.inRange(hsv_img, lower_purple, upper_purple)
    #mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)     n.b : for opencv 3.x users
    
    # To find contours OpenCv 3.x user should use the line commented below 
    # _,contours, _ = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contours, _ = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # code that creates an empty arrya to store the centroid of mass
    center = int(Height/2), int(Width/2)
    
    # if statement below gets exec if contours is found
    if len(contours) > 0:
        
        # code that gets the largest contour and its center
        c = max(contours, key=cv2.contourArea)                 # code gets the max value of the sorted contours by area.
        (x, y), radius = cv2.minEnclosingCircle(c)
        M = cv2.moments(c)
        try:
            center = (int(M['m10'] / M['m00']), int(M['m01'] / M['m00']))
        except:
            center = int(Height/2), int(Width/2)
        
        # Takes only contours that have larger than 18 pixel radius
        if radius > 18:
            # Draw Circle and leace the last center creating a trail
            cv2.circle(frame, (int(x), int(y)), int(radius), (0,0,255), 2)
            cv2.circle(frame, center, 5, (0, 255, 0), -1)
            
    # Code that logs the center points
    points.append(center)
            
    # Code that loops over the set of tracked points with radius pixel greater than 18
    if radius > 18:
        for i in range(1, len(points)):
            try:
                cv2.line(frame, points[i - 1], points[i], (0, 255, 0), 2)
            except:
                pass
                    
        # make frame count to be equals to zero
        frame_count = 0
                
    else:
        # code that count the frames
        frame_count += 1
                
        # code that del the trails everytime our object has been out for 10 frames
        if frame_count == 10:
            points = []
            # code that clears our trail once frame_count reaches 10
            frame_count = 0
                    
                    
    # code that turns image into a mirror like view
    frame = cv2.flip(frame, 1)
    cv2.imshow('Object OR Ball Tracker', frame)
            
    if cv2.waitKey(1) == 13:
        break
                
cap.release()
cv2.destroyAllWindows()
                    
```
